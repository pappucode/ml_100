```python
# Initialize an empty DataFrame
df = pd.DataFrame()

# API key and base URL
api_key = "0ec08320a961bef80f2cfd41b70e0386"
base_url = "https://api.themoviedb.org/3/movie/top_rated"

# Make the first request to get total_pages
response = requests.get(base_url, params={"api_key": api_key, "page": 1})
if response.status_code == 200:
    data = response.json()
    total_pages = data.get("total_pages", 1)  # Get total pages from the response
else:
    print(f"Error: Failed to fetch data, Status Code: {response.status_code}")
    total_pages = 0

# Loop through all available pages
for page in range(1, total_pages + 1):
    response = requests.get(base_url, params={"api_key": api_key, "page": page})
    if response.status_code == 200:
        # Parse the response and append data to the DataFrame
        temp_df = pd.DataFrame(response.json()['results'])[["id", "title", "overview", "release_date", "popularity", "vote_average", "vote_count"]]
        df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"Error: Failed to fetch page {page}, Status Code: {response.status_code}")
        break

# # Save the data to a CSV file (optional)
# df.to_csv("top_rated_movies.csv", index=False)

# # Display the first few rows
# print(df.head())
```
