```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df = pd.read_csv('datasets/train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train_trf, y_train)
y_pred = clf.predict(X_test_trf)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
mi = MissingIndicator()

mi.fit(X_train)
```

```python
# Provide result that age has missing values
mi.features_ 
```

```python
X_train_missing = mi.transform(X_train)
```

```python
X_train_missing
```

```python
X_test_missing = mi.transform(X_test)
```

```python
X_test_missing
```

```python
X_train['Age_NA'] = X_train_missing 
```

```python
X_test['Age_NA'] = X_test_missing 
```

```python
X_train
```

```python
X_test
```

```python
si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
clf = LogisticRegression()

clf.fit(X_train_trf2, y_train)

y_pred = clf.predict(X_test_trf2)

accuracy_score(y_test, y_pred)
```

```python
# Here missing indicator class given by Default.
si = SimpleImputer(add_indicator=True)
```

```python
X_train_trf3 = si.fit_transform(X_train)
X_test_trf3 = si.transform(X_test)
```

```python
clf = LogisticRegression()

clf.fit(X_train_trf3,y_train)

y_pred = clf.predict(X_test_trf3)

accuracy_score(y_test,y_pred)
```
